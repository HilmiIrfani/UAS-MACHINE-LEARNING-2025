```python
import matplotlib.pyplot as plt
import cv2
import os
```

```python
# Set path dataset
path = "/content/PlantVillage-Dataset/raw/color"
# Daftar kelas
classes = os.listdir(path)
print(f"Jumlah Kelas: {len(classes)}")
print(f"Kelas: {classes}")
```

```python
import cv2
import matplotlib.pyplot as plt

# Tampilkan gambar dari kelas pertama
sample_class = classes[0]
sample_images = os.listdir(os.path.join(path, sample_class))
```

```python
from collections import Counter
class_distribution = Counter([folder for folder in os.listdir(path)])
print("Class Distribution: ", class_distribution)
```

```python
# Load dan tampilkan gambar pertama
sample_image_path = os.path.join(path, sample_class, sample_images[0])
image = cv2.imread(sample_image_path)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title(f"Contoh Gambar: {sample_class}")
plt.axis("off")
plt.show()
```

```python
from collections import Counter

# Hitung jumlah gambar per kelas
class_counts = {cls: len(os.listdir(os.path.join(path, cls))) for cls in classes}
print("Distribusi Kelas:")
for cls, count in class_counts.items():
    print(f"{cls}: {count} gambar")
```

```python
# Visualisasi distribusi
plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=90)
plt.title("Distribusi Gambar per Kelas")
plt.show()
```

```python
import numpy as np

# Resize gambar
resized_images = []
labels = []

for cls in classes:
    class_path = os.path.join(path, cls)
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        img = cv2.imread(img_path)
        img_resized = cv2.resize(img, (128, 128))
        resized_images.append(img_resized)
        labels.append(cls)
```

```python
# Convert ke numpy array
resized_images = np.array(resized_images)
labels = np.array(labels)

print(f"Shape Gambar: {resized_images.shape}")
print(f"Jumlah Label: {len(labels)}")
```

```python
def extract_color_histogram(image):
    hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    hist = cv2.normalize(hist, hist).flatten()
    return hist

# Ekstrak fitur histogram
hist_features = [extract_color_histogram(img) for img in resized_images]

print(f"Shape Fitur Histogram: {np.array(hist_features).shape}")
```

```python
from sklearn.model_selection import train_test_split

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    hist_features, labels, test_size=0.3, random_state=42, stratify=labels
)

print(f"Jumlah Data Training: {len(X_train)}")
print(f"Jumlah Data Testing: {len(X_test)}")
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Encode labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Inisialisasi dan latih model
svm_model = SVC(kernel="linear", C=1.0, random_state=42)
svm_model.fit(X_train, y_train_encoded)
```

```python
# Prediksi pada data uji
y_pred = svm_model.predict(X_test)
```

```python
# Evaluasi akurasi
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f"Akurasi Model SVM: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_encoded, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
report = classification_report(y_test_encoded, y_pred, target_names=label_encoder.classes_)
print(report)
```

```python
from sklearn.model_selection import GridSearchCV

# Grid Search untuk optimasi
param_grid = {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "rbf"]
}
grid_search = GridSearchCV(SVC(), param_grid, cv=5)
grid_search.fit(X_train, y_train_encoded)

# Model terbaik
print(f"Best Parameters: {grid_search.best_params_}")
best_model = grid_search.best_estimator_

# Evaluasi ulang
y_pred_optimized = best_model.predict(X_test)
accuracy_optimized = accuracy_score(y_test_encoded, y_pred_optimized)
print(f"Akurasi Model Terbaik: {accuracy_optimized * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Prediksi dengan model terbaik
y_pred_optimized = best_model.predict(X_test)

# Confusion Matrix
conf_matrix_optimized = confusion_matrix(y_test_encoded, y_pred_optimized)
sns.heatmap(conf_matrix_optimized, annot=True, fmt="d", cmap="Blues",
            xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix - Optimized Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
report_optimized = classification_report(y_test_encoded, y_pred_optimized, target_names=label_encoder.classes_)
print("Classification Report - Optimized Model:\n", report_optimized)
```

```python
import cv2
import numpy as np

# Fungsi untuk ekstraksi fitur tekstur sederhana (mean, variance, dan edge density)
def extract_texture_features_opencv(image):
    # Konversi ke grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Mean (rata-rata intensitas piksel)
    mean_intensity = np.mean(gray_image)

    # Variance (variasi intensitas piksel)
    variance_intensity = np.var(gray_image)

    # Edge density menggunakan Sobel
    edges = cv2.Canny(gray_image, 100, 200)
    edge_density = np.sum(edges) / (edges.shape[0] * edges.shape[1])

    return [mean_intensity, variance_intensity, edge_density]

# Ekstraksi fitur tekstur untuk semua gambar
texture_features_opencv = np.array([extract_texture_features_opencv(img) for img in resized_images])

print(f"Shape Fitur Tekstur: {texture_features_opencv.shape}")

```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Gabungkan histogram warna dan fitur tekstur
combined_features_opencv = np.hstack((hist_features, texture_features_opencv))

# Split data ulang
X_train_comb, X_test_comb, y_train_comb, y_test_comb = train_test_split(
    combined_features_opencv, labels, test_size=0.3, random_state=42, stratify=labels
)

# Encode label
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_comb)
y_test_encoded = label_encoder.transform(y_test_comb)

# Latih model ulang dengan fitur gabungan
svm_model_comb = SVC(kernel="linear", C=1.0, random_state=42)
svm_model_comb.fit(X_train_comb, y_train_encoded)

# Evaluasi ulang
y_pred_comb = svm_model_comb.predict(X_test_comb)
accuracy_comb = accuracy_score(y_test_encoded, y_pred_comb)
print(f"Akurasi Model dengan Fitur Gabungan: {accuracy_comb * 100:.2f}%")
```
